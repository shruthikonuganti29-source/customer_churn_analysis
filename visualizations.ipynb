```python
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3

conn = sqlite3.connect("churn.db")
```

```python
query = """
SELECT Churn, COUNT(*) AS customers
FROM customers
GROUP BY Churn
"""

df1 = pd.read_sql(query, conn)

df1.plot(kind="bar", x="Churn", y="customers", legend=False)
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Customers")
plt.show()
```

```python
query = """
SELECT Contract, COUNT(*) AS customers
FROM customers
GROUP BY Contract
"""

df2 = pd.read_sql(query, conn)

df2.plot(kind="pie", y="customers", labels=df2["Contract"], autopct="%1.1f%%")
plt.title("Customers by Contract Type")
plt.ylabel("")
plt.show()
```

```python
query = """
SELECT Contract, AVG(MonthlyCharges) AS avg_charge
FROM customers
GROUP BY Contract
"""

df3 = pd.read_sql(query, conn)

df3.plot(kind="line", x="Contract", y="avg_charge", marker="o", legend=False)
plt.title("Average Monthly Charges by Contract")
plt.xlabel("Contract")
plt.ylabel("Charges")
plt.show()
```

```python
query = """
SELECT PaymentMethod, COUNT(*) AS customers
FROM customers
GROUP BY PaymentMethod
ORDER BY customers DESC
"""

df4 = pd.read_sql(query, conn)

df4.plot(kind="bar", x="PaymentMethod", y="customers", legend=False)
plt.title("Customers by Payment Method")
plt.xticks(rotation=45)
plt.show()
```
